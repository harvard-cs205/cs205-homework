```python
import findspark
findspark.init()
import pyspark
import json
sc = pyspark.SparkContext(appName="Spark1")
```

```python
fp = open('source.csv', 'r')
c = fp.read().split('\n')[:-1]
fp.close()
print c
```

```python
movie_dict = {}
graph = {}
for temp in c:
    temp_name = temp.split(" "," ")
    movie_name = temp_name[0][1:]
    character_name = temp_[1][:-2]
    if character_name not in movie_dict:
        movie_dict[character_name] = []
    if character_name not in graph: 
        graph[character_name] = []
        graph[character_name].append(character_name)
    if movie in movie_dict and character_name not in movie_dict[movie]:
        movie_dict[movie].append(character_name)
        
    
```
